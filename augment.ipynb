```python
import re

def check_prefix(string):
    return bool(re.match(r"^\d+%https://", string))

prc_lines=[]
with open(txt, 'r') as fi:
    lines=fi.readlines()
    for ln in lines:
        ln=ln.strip()
        if not check_prefix(ln):
            prc_lines[-1]=prc_lines[-1]+" "+ln
        else:
            prc_lines.append(ln)

with open("process_items.txt", 'w') as fi:
    fi.writelines([ln+"\n" for ln in prc_lines])
```

Because districts of HCM and HN take up a large portion of the data, it's a good practice to separate the two cities into smaller locations.

```python
chk_res=[]
add_dic={}

all_add=["tp. hcm", "hà nội"]
for res in res_inf:
    if "-" in res[3] or "B0" in res[3] or "B6" in res[3]:
        continue
    
    if res[0] not in chk_res:
        chk_res.append(res[0])
        if "HCM" in res[3] or "Hà Nội" in res[3]:
            add_dic[res[3].split(",")[-2].strip().lower()]=res[3].split(",")[-1].strip().lower()
        else:
            all_add.append(res[3].split(",")[-1].strip().lower())

all_add+=list(add_dic.keys())
all_add=list(set(all_add))
all_add.sort()
print(all_add)
```

```python
def get_districts(city, dic):
    dis=[]
    for d, c in dic.items():
        if city==c:
            dis.append(d)
    return dis+[city, ""]

def get_food_classes(food, dic):
    dis=[]
    if food=="":
        return []
    
    if "deal" in food or "giảm giá" in food or "đang giảm" in food:
        return ["khuyến mãi"]
    
    if food in dic:
        return [food, dic[food]]
    else:
        return [food]

```

Create two dicts that store data about restaurants and food categories

```python
# Ten_quan_an - Dia_chi - ~Dia_chi - Loai_mon_an - ~Loai_mon_an - Ten_mon_an
# Loai_mon_an - Ten_mon_an
all_prs=[]
res_dic={}
men_dic={}
for res, prd in prc_lines[:]:
    if "- " in res[3].split(",")[-1] or "B0" in res[3].split(",")[-1] or "B6" in res[3].split(",")[-1]:
        continue
    
    # transfer string to lower
    res=[res[0].lower(), res[1].lower(), res[2].lower(), res[3].lower()]
    prd=[prd[0].lower(), prd[1].lower(), prd[2].lower()]

    # get the name of place
    nm=res[2]
    if nm not in res_dic:
        res_dic[nm]={}
    
    # get the city => if the city is either HCM or HN, get the district
    # get cities that the place is not belong to
    ad=res[3].split(",")[-1].strip()
    if "hcm" in ad:
        res_dic[nm]["dia_chi"]=[res[3].split(",")[-2].strip(), ad]
    else:
        res_dic[nm]["dia_chi"]=[res[3].split(",")[-2].strip()]
    
    res_dic[nm]["not_dia_chi"]=list(set(all_add).difference(get_districts(ad, add_dic)))

    # get class of food
    mon_an=re.sub(r'[^a-zA-Záàảãạăắằẳẵặâấầẩẫậéèẻẽẹêếềểễệíìỉĩịóòỏõọôốồổỗộơớờởỡợúùủũụưứừửữựýỳỷỹỵđ\s]', '', prd[1])
    mon_an=re.sub(r'\s+', ' ', mon_an)
    mon_an=mon_an.strip()

    if "loai_mon_an" not in res_dic[nm]:
        res_dic[nm]["loai_mon_an"]=[]
    res_dic[nm]["loai_mon_an"].extend(get_food_classes(prd[-1], cls_dic))

    # if prd[-1] in chk_cls:
    if prd[-1] not in men_dic:
        men_dic[prd[-1]]=[]
    men_dic[prd[-1]].append(mon_an)

    # get food
    if "mon_an" not in res_dic[nm]:
        res_dic[nm]["mon_an"]=[]
    res_dic[nm]["mon_an"].append(mon_an)

    # get all food
    if prd[1] not in all_prs:
        all_prs.append(prd[1])

for res in res_dic.keys():
    res_dic[res]["loai_mon_an"]=list(set(res_dic[res]["loai_mon_an"]))
    res_dic[res]["mon_tuong_tu"]=[]
    for cl in res_dic[res]["loai_mon_an"]:
        if cl in chk_cls and cl in men_dic:
            res_dic[res]["mon_tuong_tu"].extend(men_dic[cl])
    
    res_dic[res]["mon_tuong_tu"]=list(set(res_dic[res]["mon_tuong_tu"]).difference(res_dic[res]["mon_an"]))

# for res in res_dic.keys():
#     print(res_dic[res])

    
```
